```python
from model import NewModel
```

```python
label_config = """
<View>
  <Image name="image" value="$image" />
  <RectangleLabels name="label" toName="image">
    <Label value="person" />
    <Label value="car" />
    <Label value="dog" />
  </RectangleLabels>
</View>
"""
```

```python
backend = NewModel(label_config=label_config)
```

```python
import base64

def to_datauri(path):
    with open(path, "rb") as f:
        return "data:image/jpeg;base64," + base64.b64encode(f.read()).decode()

task = {
    "data": {
        "image": to_datauri("osaka.jpg")
    }
}

pred = backend.predict([task])
pred
```
